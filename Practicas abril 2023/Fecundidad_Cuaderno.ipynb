## **2.3 Limpieza de datos**

Este apartado consistirá en eliminar valores perdidos o imputarlos. 


```python
print(fecundidad.isnull().sum())  #suma el numero de valores perdidos que tiene cada una de las variables
```

```python
fecundidad_11_na = fecundidad.dropna(thresh=fecundidad.shape[1]-10) #elimina las observaciones que tienen 11 valores perdidos, es decir, que tienen valor perdido en cada variable con valores perdidos
print(fecundidad_11_na.isnull().sum())  #suma el numero de valores perdidos que tiene cada una de las variables
```

Primero eliminamos las observaciones que tienen valores perdidos en todas las variables con valores perdidos. En este caso, indicamos en el código un 10, pero Python entiende que hay que eliminar las observaciones que tienen 11 valores perdidos en una observación. <br>
Posteriormente, vamos a eliminar las observaciones que tienen valores perdidos en las variables TIPOUNION, NACPAR, ESTUDIOSPAR, SATISRELAC y TEMPRELA conjuntamente, dado que, dichas observaciones tienen valores perdidos para todas esas variables. 

```python
fecundidad_11_5_na = fecundidad_11_na.dropna(subset=['TIPOUNION']) #elimina las observaciones que tienen valores perdidos en la variable TIPOUNION. No hace falta poner las otras 4 variables porque esas observaciones coinciden
print(fecundidad_11_5_na.isnull().sum())  #suma el numero de valores perdidos que tiene cada una de las variables
```

Una vez se han eliminado las observaciones que presentaban valores perdidos conjuntamente en las variables TIPOUNION, NACPAR, ESTUDIOSPAR Y SATISRELAC, procederemos a eliminar de la base de datos las variables INTENCASARSE, M_NOHIJOS1, M_NOHIJOS2 Y M_NOHIJOS3, pues tienen más missing que observaciones con información y, por consiguiente, no son de la utilidad para el estudio. 

```python
fecundidad_11_5_na_vars=fecundidad_11_5_na.drop(columns=['INTENCASARSE','M_NOHIJOS1','M_NOHIJOS2','M_NOHIJOS3'])
print(fecundidad_11_5_na_vars.isnull().sum())  #suma el numero de valores perdidos que tiene cada una de las variables
```

A continuación, los valores perdidos de la variable PRACTICANTE serán imputados y, al ser una variable cualitativa, se utilizará la moda para ello. <br>

```python
fecundidad_11_5_na_vars.PRACTICANTE.value_counts(normalize=True) #Para ver cual es la categoría modal
```

```python
fecundidad_11_5_na_vars['PRACTICANTE'].fillna(fecundidad_11_5_na_vars['PRACTICANTE'].mode()[0], inplace=True) #Remplaza los valores perdidos por la moda
```

```python
print(fecundidad_11_5_na_vars.isnull().sum())  #comprobación de que se han imputado los datos
```

Por último, el objetivo será eliminar las observaciones con valor perdido en la variable EMBACT (está usted embarazada actualmente). En esta ocasión, no trataremos de imputar las observaciones dado que, al tratarse de una variable muy comprometedora, los valores faltantes no se han perdido, sino que los entrevistados no han querido responder por elección propia.

```python
fecundidad_11_5_na_vars_final = fecundidad_11_5_na_vars.dropna(subset=['EMBACT'])
```

```python
print(fecundidad_11_5_na_vars_final.isnull().sum())  #comprobación de que se han eliminado dichas observaciones
```

Finalmente, se ha conseguido eliminar o imputar todos los valores perdidos de la base de datos. 

```python
fecundidad_sin_na=fecundidad_11_5_na_vars_final #El fichero de datos sin perdidos se llamará fecundidad_sin_na
fecundidad_sin_na
```

## **2.4 Análisis estadístico (Análisis de Correspondencias)**

En este apartado nos vamos a enfocar en detectar asociación entre variables para reducir la dimensión. 
También podemos reducir la dimensión agrupando las variables N_HERMANOS y N_HERMANAS por ejemplo. <br>
Dado que el conjunto de datos cuenta con muchas variables, se van a agrupar en diferentes grupos de variables. El primero de los grupos lo componen las variables COD_CCAA, EC, NAC, TIPOHO, INGREHOG_INTER y DIFICULTAD. A esta agrupación se le denominará datos biográficos. <br>


```python
datos_biograficos = fecundidad_sin_na.iloc[:, 2:8]  
datos_biograficos
```

En segundo lugar, se van a agrupar las variables TIPOUNION, NACPAR, ESTUDIOSPAR	SATISRELAC, SITSENTI y TEMPRELA. Este subgrupo recibirá el nombre de pareja actúal

```python
pareja_actual = fecundidad_sin_na.iloc[:, [12,13,14,15,21,22]]  
pareja_actual
```

El resto de variables no forman grupos lo suficientemente grandes como para que sea interesante realizar un análisis de correspondencias por separado. Por consiguiente, se tendrán en cuenta el resto de variables forma individual, es decir, se va a realizar un análisis de correspondencias del resto de variables.  

```python
resto = fecundidad_sin_na.iloc[:, [1, 8, 9, 10, 15, 16 , 17, 18, 19, 22]]  
resto
```

**Análisis de correspondencias del subgrupo datos biográficos:**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix

le = LabelEncoder()

for col in datos_biograficos.columns:
    if datos_biograficos[col].dtype == 'object':
        datos_biograficos[col] = le.fit_transform(datos_biograficos[col])
```

```python
corr_matrix = pd.DataFrame(index=datos_biograficos.columns, columns=datos_biograficos.columns)

for col1 in datos_biograficos.columns:
    for col2 in datos_biograficos.columns:
        corr_matrix.loc[col1, col2] = np.sum((datos_biograficos[col1] == datos_biograficos[col2]) & (datos_biograficos[col1] != -1))

corr_matrix = corr_matrix.astype('float')

for col in corr_matrix.columns:
    corr_matrix[col] = corr_matrix[col] / np.sum(corr_matrix[col])
```

```python
pca = PCA(n_components=2)

pca.fit(corr_matrix)

pca_matrix = pca.transform(corr_matrix)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))  # se define el tamaño de la figura

plt.scatter(pca_matrix[:,0], pca_matrix[:,1])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

variables = list(datos_biograficos.columns)

for i, variable in enumerate(variables):
    plt.annotate(variable, (pca_matrix[i,0], pca_matrix[i,1]), ha='right')
    
plt.show()
```

Las variables DIFICULTAD e INGREHOG_INTER, están muy correlacionadas. Por consiguiente, para reducir la dimensión se utilizará únicamente la variable DIFICULTAD.

## **2.6 Librería Pandas-Profiling** 


Por último, antes de finalizar el análisis exploratorio de los datos, merece la pena destacar que Python ofrece una librería llamada Pandas-Profiling que realiza de forma sistemática y rápida un EDA de cualquier conjunto de datos.
<br>

**Soporte teórico:**

Pandas profiling es una librería de Python que permite generar informes de perfilado de datos para explorar y entender conjuntos de datos de forma rápida y sencilla. <br>

La librería pandas profiling se basa en la librería pandas de Python y proporciona una interfaz fácil de usar para generar un informe de perfilado de datos en formato HTML que incluye información detallada sobre las variables (columnas) en un conjunto de datos, como:

El tipo de datos (numérico, categórico, booleano, etc.)
El número de valores únicos y los valores más comunes
El porcentaje de valores faltantes
Estadísticas descriptivas (media, desviación estándar, valores mínimo y máximo, etc.)
Correlaciones entre variables
Histogramas, diagramas de cajas y otros gráficos para visualizar la distribución de los datos
Además, pandas profiling también permite personalizar el informe de perfilado de datos, para que los usuarios puedan ajustar el nivel de detalle y la complejidad del informe en función de sus necesidades.

En general, pandas profiling es una herramienta muy útil para explorar y entender conjuntos de datos de forma rápida y sencilla, lo que puede ahorrar mucho tiempo y esfuerzo en el análisis de datos.

## **3.1 Árbol de clasificación**

**Soporte teórico: <br>**
Los árboles de clasificación constituyen una herramienta útil para la predicción de variables cualitativas de una manera sencilla y sin asunciones teóricas sobre los datos. <br>
Estos representan una segmentación de los datos a partir de una serie de reglas simples, que se van aplicando de forma jerárquica y secuencial. De esta forma, se obtienen una serie de segmentos (llamados nodos) que contienen subconjuntos de la muestra. <br>
Una vez obtenida la segmentación “óptima”, entendiendo así aquella que da lugar a nodos con comportamiento homogéneo respecto a la variable dependiente y heterogéneos entre sí, se asigna un valor de predicción a aquellos nodos que no tienen sucesores (y que reciben el nombre de hojas) de forma que todas las observaciones pertenecientes a dicha hoja serán predichas a partir de dicho valor. <br>

A modo de inicialización de este análisis, se eliminarán las variables que no deben estar debido a su naturaleza, dado que, si se incluyen pueden llevar a resultados erróneos en el árbol de clasificación. 

```python
X = fecundidad_sin_na.drop('EMBANT', axis=1)
X = X.drop('IDEN', axis=1)
X = X.drop('COD_CCAA', axis=1)
Y = fecundidad_sin_na['EMBANT']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)  #Dividimos la muestra en train y test
# 8794*0.7~=6155 => Tamaño del conjunto de datos train  
# 8794*0.3~=2639 => Tamaño del conjunto de datos test
```

**Busqueda de los hiperparámetros:**


*   **max_depth**: Este parámetro especifica la profundidad máxima del árbol. Un valor más alto de max_depth puede conducir a un modelo más complejo que se ajusta demasiado a los datos de entrenamiento, mientras que un valor más bajo puede llevar a un modelo insuficientemente complejo que no se ajusta bien a los datos. Por lo tanto, es importante ajustar este parámetro para encontrar un equilibrio entre la complejidad y la capacidad de generalización del modelo.
*   **min_samples_split**: Este parámetro especifica el número mínimo de muestras requeridas para dividir un nodo interno. Un valor más alto de min_samples_split previene la formación de particiones con pocas muestras, lo que puede conducir a una mayor generalización del modelo. Por otro lado, un valor más bajo de min_samples_split permitiría formar particiones con un número menor de muestras, lo que podría conducir a un sobreajuste en el modelo.

En resumen, max_depth controla la complejidad del modelo, mientras que min_samples_split controla la cantidad de datos requeridos para la formación de nuevas particiones. Ajustar estos hiperparámetros adecuadamente puede ayudar a obtener un modelo de árbol de clasificación más preciso y generalizable.

**Representación del árbol de clasificación:**

*   **Índice de Gini**: es una medida de impureza utilizada en la construcción de árboles de decisión y árboles de clasificación. En el contexto de un árbol de clasificación en Python, gini se refiere a la impureza de un nodo, que mide la probabilidad de que una muestra aleatoria sea clasificada incorrectamente en ese nodo. Un valor de 0 indica que el nodo es completamente puro (todas las muestras en el nodo pertenecen a la misma clase), mientras que un valor de 0.5 indica que el nodo es igualmente impuro para todas las clases. 
*   **samples**: Este parámetro indica el número de muestras (observaciones) en un nodo dado del árbol de clasificación.
*  **value**: Este parámetro muestra la distribución de clases en un nodo dado. Por ejemplo, si un nodo tiene 30 muestras y 20 de ellas pertenecen a la clase A y 10 a la clase B, el valor se mostrará como [20, 10], indicando la distribución de las clases en ese nodo.
*   **class**: Este parámetro muestra la clase mayoritaria en un nodo dado. Por ejemplo, si un nodo tiene 30 muestras y 20 de ellas pertenecen a la clase A y 10 a la clase B, la clase se mostrará como A. Este parámetro es útil para interpretar la clasificación final de una muestra en el árbol de clasificación.

Asimismo, la intensidad del color se debe a lo homogéneos o heterogéneos que son cada uno de los nodos, por lo que, colores más oscuros estarán relacionados con nodos extremos en cuanto a las probabilidades a posteriori, mientras que, colores más claros estarán relacionados con nodos similares en cuanto a las probabilidades a posteriori. <br>

**Evaluación del modelo:**


*   **La tasa de acierto (accuracy)** es una medida comúnmente utilizada en los árboles de clasificación que indica la proporción de predicciones correctas en relación al total de predicciones realizadas.
*  **El kappa** es una medida de concordancia que evalúa la precisión de la clasificación de un modelo ajustado en relación a una clasificación de referencia. El kappa toma en cuenta tanto la proporción de acuerdos como la proporción de desacuerdos que se esperan por casualidad. Valores mayores de kappa indican una mayor concordancia entre el modelo y la clasificación de referencia.


*   **La sensibilidad** es la proporción de muestras positivas que fueron correctamente identificadas por el modelo. Una alta sensibilidad indica que el modelo es capaz de identificar la mayoría de las muestras positivas.
*  **La especificidad** es la proporción de muestras negativas que fueron correctamente identificadas por el modelo. Una alta especificidad indica que el modelo es capaz de identificar la mayoría de las muestras negativas.




**Entrenamiento:**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import confusion_matrix

y_pred = arbol.predict(X_train) #Se calculan los valores predichos para train
#Cálculo de las métricas
accuracy = accuracy_score(y_train, y_pred)
kappa = cohen_kappa_score(y_train, y_pred)
cm = confusion_matrix(y_train, y_pred)
tn, fp, fn, tp = confusion_matrix(y_train, y_pred).ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

#Salida por pantalla de las métricas
print("ENTRENAMIENTO")
print(cm) #Matriz de confusión
print("Accurazy: ", accuracy)
print("KAPPA: ", kappa)
print("SENSIBILIDAD: ", sensitivity)
print("ESPECIFICIDAD: ", specificity)

```

A continuación, se representará la **Curva ROC:** 

Las medidas derivadas de la matriz de confusión presentan el inconveniente de depender del punto de corte que clasifica a las observaciones como evento y no evento. Por tanto, a continuación, se obtendrá la curva ROC para contar con otra herramienta que permita evaluar el modelo. Esta curva permite obtener una medida de evaluación sin necesidad de fijar un punto de corte. Para ello, se representa en el eje Y la sensibilidad y en el eje X 1-Especificidad, que se han obtenido a partir de todos los posibles puntos de corte y las correspondientes clasificaciones. <br>
Cuanto más cóncava sea la curva ROC, mejor será el modelo que se está evaluando. Dado que en ocasiones es difícil visualizar que curva es más cóncava, se define el área bajo la curva (area under the curve=AUC). Un modelo perfecto tendrá un AUC de 1, mientras que el peor modelo tendrá un AUC de 0.5.

```python
#Calcula la tasa de verdaderos positivos (TPR), la tasa de falsos positivos (FPR)
from sklearn.metrics import roc_curve
y_probs = arbol.predict_proba(X_train)[:, 1] # obtener la probabilidad de la clase positiva
fpr, tpr, thresholds = roc_curve(y_train, y_probs)

#Calcular el área para la curva ROC
from sklearn.metrics import auc
roc_auc = auc(fpr, tpr)

#Crea el gráfico de la curva ROC
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, label='AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1-Especificidad')
plt.ylabel('Sensibilidad')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()
```

Una vez obtenido el valor del **AUC=0.85**, se puede determinar que el modelo es de muy buena calidad.

## **3.2 Random Forest (RF)**

**Soporte teórico:** <br>
Random Forest (RF) es un algoritmo de aprendizaje automático supervisado utilizado para la clasificación y regresión. Es una extensión de los árboles de decisión en el que se utiliza una combinación de varios árboles de decisión para mejorar el rendimiento y la precisión del modelo. <br>
El funcionamiento de un Random Forest consiste en crear múltiples árboles de decisión aleatorios y luego combinar sus resultados para llegar a una decisión final. En lugar de entrenar un solo árbol de decisión, RF genera una serie de árboles de decisión aleatorios utilizando diferentes muestras de datos y características para cada árbol. Cada árbol toma una decisión basada en sus características aleatorias y, finalmente, se combina el resultado de todos los árboles para tomar la decisión final. <br>
La ventaja de Random Forest sobre los árboles de decisión es que reduce el sobreajuste y aumenta la precisión y estabilidad del modelo.
