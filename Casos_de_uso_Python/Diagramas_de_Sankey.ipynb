##**Diagrama de Sankey**

Este tipo de diagramas tratan de mostrar gráficamente, con la anchura de sus flujos, la cantidad que va desde el origen de una variable al destino de otra. En este notebook se plantean 4 formas de representar un diagrama de Sankey.

La información la optenemos en las estadísticas experimentales del INE (https://www.ine.es/experimental/movilidad/experimental_em4.htm#tablas_resultados), donde nos quedaremos con los datos del 25 de diciembre de 2021, ubicados en el archivo *Tabla 3.7 Movilidad Estacional-Población que pernocta en una CCAA diferente de la de residencia*. Aunque en nuestro dataset hemos cambiado los datos que se mostraban en horizontal para mostrarlos en vertical, para así poder leerlos en Colab como a nosotros nos interesa. 

En este primer ejemplo, vamos a representar un diagrama de Sankey con el paquete **pySankey**. Lo primero que haremos será instalar el paquete.

###**Ejemplo 2**
En nuestro segundo ejemplo, vamos a utilizar el mismo dataset y vamos a seguir la misma estructura, con la salvedad de que en esta ocasión utilizaremos el paquete **matplotlib**.

```python
import matplotlib.pyplot as plt
```

En esta ocasión, cuando aplicamos la función sankey(), le indicamos que datos queremos que se posicionen a la izquierda y a la derecha, así como que los fujos se correspondan a los tamaños correspondientes con leftWeight y rightWeight. 
Además, establcemos los tamaños en los que queremos que se visualice el diagrama. Por último, podemos guardar el diagrama como una imagen. 

```python
colors = {
    "Andalucía": "#EE82EE",
    "Aragón": "#D8BFD8",
    "Asturias, Principado de": "#FFA54F",
    "Balears, Illes": "#7FFFD4",
    "Canarias": "#f78c1b", 
    "Cantabria": "#E3CF57",
    "Castilla y León": "#CDB79E",
    "Castilla-La Mancha": "#FF4040",
    "Cataluña": "#5F9EA0",
    "Comunitat Valenciana": "#8EE5EE",
    "Extremadura": "#ED9121",
    "Galicia": "#66CD00",
    "Madrid, Comunidad de": "#FF7F50",
    "Murcia, Región de": "#A9A9A9",
    "Navarra,Comunidad Foral de": "#FF6A6A",
    "País Vasco": "#F08080",
    "Rioja, La": "#8470FF",
    "Ceuta": "#DDA0DD",
    "Melilla": "#CDCD00"
}
sankey(
    left=data["CCAA_res"], right=data["CCAA_pec"], 
    leftWeight= data["Personas"], rightWeight=data["Personas"],    
    aspect=20, colorDict=colors, fontsize=12)

# Obtener la referencia de la cifra actual
fig = plt.gcf()

# Tamaño de la figura
fig.set_size_inches(6, 6)

# Color de fondo del gráfico
fig.set_facecolor("w")

# Altura y anchura de la figura
fig.set_figheight(12)
fig.set_figwidth(15)
fig.show()

# Guardar el diagrama
fig.savefig("sankey_ccaa.png")
```

###**Ejemplo 3**
En este tercer ejemplo vamos a utilizar la libreria **plotly**, el cual permite hacer visualizaciones interactivas. Dado que ya no siguen la misma estructura, vamos a cargar de nuevo el dataset ***Movilidad_estacional***. 
Para que funcione correctamente, en el dataset se ha introducido la columna source, el cual nos mostrará la posición de origen de la variable (CCAA), y la columna target, que nos señala la posición de la variable de destino. 

```python
# Preparación de los datos 
import pandas as pd
import numpy as np
# Visualización
from plotly.offline import iplot
import plotly.graph_objs as go
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data=pd.read_excel('Movilidad_estacional.xlsx')
data.head()
```

Creamos una lista con el nombre de las variables, que en este caso son las Comunidades Autónomas. Como tenemos las mismas variables de origen que de destino, es decir, tenemos las mismas variables a la izquierda y a la derecha, creamos la variable label, la cual contiene tanto las variables de un lado y de otro, y como son las mismas, pues solo hay que sumar la lista de ccaa consigo misma.

```python
ccaa = ["Andalucía","Aragón","Asturias, Principado de","Balears, Illes",
"Canarias", "Cantabria","Castilla y León","Castilla-La Mancha",
"Cataluña","Comunitat Valenciana","Extremadura","Galicia",
"Madrid, Comunidad de","Murcia, Región de","Navarra,Comunidad Foral de",
"País Vasco","Rioja, La","Ceuta","Melilla"]
label = ccaa + ccaa
```

Vamos a aplicar una secuencia de colores a nuestras variables, para ello tenemos que cargar **Plotly Express**, el cual se encarga de aplicar esta secuencia. Dado el número de variables, aplicaremos la secuencia Alphabet, que tiene 25 colores, pero también podríamos aplicar otras como Dark24 o Light24, ambas con 24 colores, la primera con tonos más oscuros y la segunda con tonos más claros. 
Además, para relacionar nuestras variables con los colores, indicamos que queremos con color nuestra variable que contiene todas las variables (label), pero que queremos que tome como referencia la columna de CCAA_res, es decir, los colores vendrán marcado por las variables en esa columna. 
Para contruir nuestro diagrama nos hacen falta las columnas de *source* y *target*, las cuales determinan la posición de nuestras variables. Así como la columna *Personas* que determina el valor de esos flujos. 

```python
import plotly.express as pex
colors = pex.colors.qualitative.Alphabet

node_colors_mappings = dict([(node,np.random.choice(colors)) for node in label])
node_colors = [node_colors_mappings[node] for node in label]
edge_colors = [node_colors_mappings[node] for node in data['CCAA_res']]

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = label,
      color= node_colors
    ),
    link = dict(
      source = data['source'].tolist(),
      target = data['target'].tolist(),
      value = data['Personas'].tolist(),
      color = edge_colors,
  ))])

fig.update_layout(title_text="Movilidad estacional por Comunidades Autonomas", font_size=10)
fig.update_layout(width=1200, height=1000)
fig.show()
```

Si nos situamos encima de los flujos podemos observar de que Comunidad Autónoma vienen las personas que luego pernoctan en otra comunidad. Siendo los ciudadanos de la Comunidad de Madrid los que más pernoctaciones hacen, siendo, además, la comunidad que más turistas nacionales recibió el 25 de diciembre de 2021.

Para descargarte el diagrama en un archivo png, como una imagen fija del diagrama, tienes que desplazarte a la parte de arriba a la derecha del gráfico y encontraras un icono de una cámara donde al pulsarlo se te descarga el archivo de la imagen fija del diagrama.
