# **MÉTODO DE NEWTON-RAPHSON**

**Objetivo:** Calcular las raíces de la ecuación $f(x)=0$ de forma aproximada.

El método de Newton-Raphson se basa en el desarrollo de Taylor de la función cuya raíz se quiere calcular. Este método permite hallar una raíz de una ecuación no-lineal si conocemos un valor numérico cercano a la raíz.\
Suponemos que $f(x)$ es una función derivable. Si $x_0$ está cerca de la solución a la función $f(x)=0$  podemos aproximarnos $f(x)$ con la tangente en $x_0$ y calcular la intersección de la recta tangente.\
La ecuación de la recta tangente en $x_0$ es:




*    $y = f'(x_0)(x - x_0) + f(x_0)$


Buscamos el punto donde $f(x_1)=0$ :


*   $0 = f'(x_0)(x_1 - x_0) + f(x_0)$

Despejando $x_1$, nos queda: 


*   $x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}$

Si queremos repetir el proceso repetidamente (el problema tendrá más de una iteración), de forma general nos quedará:

*   $x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$

## ***Ejemplo***

Los parámetros utilizados son:

*   **f** = función. Función por la cual estamos buscando la solución f(x)=0.
*   **Df** = función. Derivada de f(x).
*   **$x_0$**= número. Estimación inicial para la solución f(x)=0.
*   **tole**= número. Tolerancia. El criterio se para si $|f(x)|<\epsilon$
*   **max_iteraciones**= número entero. Número máximo de iteraciones.






```python
def newton(f,Df,x0,epsilon,max_iter): #aproximación de la solución de f(x)=0 por el método de Newton Raphson.   
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon: #si ocurre, devuelve xn. 
            print('Solución encontrada despues de',n,'iteraciones.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0: 
            print('Derivada igual a cero. No se ha encontrado solución.')
            return None
        xn = xn - fxn/Dfxn #si el número de iteraciones > max_iter (máximo de iteraciones)
    print('Excedido el número máximo de iteraciones. No se ha encontrado solución.')
    return None
```

```python
#si converge
p = lambda x: x**3 - x**2 - 1
Dp = lambda x: 3*x**2 - 2*x
approx = newton(p,Dp,1,1e-10,10)
print(approx)
```

```python
#si diverge
f = lambda x: x**(1/3)
Df = lambda x: (1/3)*x**(-2/3)
approx = newton(f,Df,0.1,1e-2,100)
```

# **Calculadora de Derivadas**

La **librería SymPy** puede usarse para realizar operaciones matemáticas complejas como derivadas en funciones en Python. La función **diff()** dentro de la biblioteca SymPy se puede utilizar para calcular la derivada de una función. 

```python
#primera derivada
import sympy as sp
x = sp.Symbol('x') 
y = x**2 + 1 
sp.diff(y,x)
```

```python
#segunda derivada
import sympy as sp 
x = sp.Symbol('x') 
y = x**3+2*x
sp.diff(y,x,x)
```

```python
#tercera derivada
import sympy as sp
x = sp.Symbol('x') 
y = x**2+5*x 
sp.diff(y,x,3)
```
