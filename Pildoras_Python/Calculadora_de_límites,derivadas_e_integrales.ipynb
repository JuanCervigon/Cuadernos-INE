La función **limit()** calcula límites.\

La función **sp.integrate** dentro de SymPy se utiliza para calcular integrales.

## **LÍMITES**

Se dice que la función $f(x)$ tiene como límite el número $L$, cuando $x$ tiende a $x_0$, si
fijado un número real positivo $ε$ , mayor que cero, existe un numero positivo $δ$ dependiente
de $ε$ , tal que, para todos los valores de $x$ distintos de $x_0$ que cumplen la condición $|x - x0| <
δ$ , se cumple que $|f(x) - L| <ε$ .


 $\lim_{x \rightarrow c } f(x) = L$image.png

 ***Ejemplos:***

**Límite cuando x tiende a cero:**

```python
from sympy import limit,Symbol 
x = Symbol('x') 
y= 1/x**2
limit(y, x, 0)
```

El límite de la función $\frac{1}{x^2}$ cuando x tiende a cero es:\
$\lim_{x \rightarrow 0 } \frac{1}{x^2} = \infty$

***Límite cuando x tiende a infinito:***

```python
from sympy import limit, oo, Symbol 
x = Symbol('x') 
y=(1+1/x)**x
limit(y, x, oo)
```

El límite de la función $(1+\frac{1}{x})^x$ cuando x tiende a infinito es:\
$\lim_{x \rightarrow \infty  } (1+\frac{1}{x})^x = e$

***Límite cuando x tiende a un número:***

```python
from sympy import limit, oo, Symbol
n = Symbol('x') 
y=1/n**n 
limit(y, n, 5)
```

El límite de la función $$\frac{1}{n^n}$$ cuando x tiende a 5 es:


$\lim_{x \rightarrow 5  } \frac{1}{n^n} = \frac{1}{3125}$

## **DERIVADAS**

Podemos decir que la derivada de un función f en un punto $x_0$ "mide" la rapidez con la que crece o decrece f en $x_0$.\
La **derivada** es el resultado de un límite y representa la recta tangente a la gráfica de la función en un punto.


La definición de la derivada es:


image.png

decimos que f es derivable en $a$ asi el límite existe y es finito.

También, una derivada se puede representar como:\
$f'(x)=\frac{dy}{dx}$

***Ejemplos:***

*La primera derivada se calcula como:*

```python
#primera derivada
import sympy as sp
x = sp.Symbol('x') 
y = x**2 + 1 
sp.diff(y,x)
```

*La segunda derivada se calcula como:*

```python
#segunda derivada
import sympy as sp 
x = sp.Symbol('x') 
y = x**3+2*x
sp.diff(y,x,x)
```

*La tercera derivada se calcula como:*

```python
#tercera derivada
import sympy as sp
x = sp.Symbol('x') 
y = x**2+5*x 
sp.diff(y,x,3)
```

# **INTEGRALES**

Sea $D$ un intervalo real cualquiera y sea $f$: $D$ $\longrightarrow$ $\mathbb R$ una función continua en $D$. Diremos que la función $F:D\longrightarrow \mathbb R$ es una función primitiva de $f$ si F es una función derivable y verifica: $F'(x)=f(x)$, para todo $x \in D$.


**Integrales primitivas:**\
Dada una función $F:D\longrightarrow \mathbb R$ continua, llamaremos integral indefinida o primitiva de f y lo denotaremos por $\int f(x) \:\: dx$, al conjunto de todas las funciones primitivas de f.

***Ejemplo:***

```python
import sympy as sp 
y=sp.cos(x) 
sp.integrate(y, x)
```

$$\int cos (x) \; dx = sin (x) +c$$

**Integrales definidas:**

La integración de funciones reales de variable real surge como herramienta para el cálculo de áreas de superficies planas, es decir:\
Sea $f :[a,b] \longrightarrow \mathbb R$  una función continua, y supongamos además, que f es no negativa, es decid, verifica $f(x)\geq 0$ para todo $x \in [a,b]$.



$$\int_{a}^{b} f(x)   \:\:  dx$$

***Ejemplo***:

```python
import sympy as sp 
x = sp.Symbol('x')
y=3*x**2  #función de densidad buscar
sp.integrate(y,(x,4,6))
```

$$\int_{4}^{6} 3x^2  \:\:  dx = 152$$
