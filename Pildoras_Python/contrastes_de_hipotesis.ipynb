# **Contraste de Hipótesis.**

El **contraste de hipótesis** es una técnica muy utilizada en inferencia estadística. Básicamente consiste en formular una hipótesis de trabajo (hipótesis nula) y enfrentarla con los resultados de un test. Si los resultados del test son compatibles con la hipótesis  nula, se acepta dicha hipótesis, pero si son incompatibles se rechaza. En este cuaderno vamos a utilizar el contraste de hipótesis para determinar si una moneda está equilibrada o no. En concreto, los conceptos que vamos a utilizar con este sencillo ejemplo son:

*   Hipótesis nula y alternativa
*   Test y Estadístico
*   Nivel de confianza, significatividad y región de rechazo
*   p-valor de un estadístico
*   Error tipo I y tipo II







### ***Hipótesis nula y alternativa.***

En primer lugar, tenemos que definir con claridad la hipótesis nula, que en este caso es “la moneda es equilibrada”. Suponemos que esa hipótesis es cierta y tenemos que enfrentarla con los resultados de un test para determinar si se ACEPTA o RECHAZA, en función de si los resultados del test con compatibles con la hipótesis nula o no. 

$\left\{
    \begin{array}{ll}
            H_0 & \mbox{ } la\mbox{ }  moneda \mbox{ } está\mbox{ }  equilibrada \\
             H_1  & \mbox{ } la\mbox{ } moneda \mbox{ } no \mbox{ } está\mbox{ } equilibrada. 
    \end{array}
\right.$

### ***Test y estadístico.***

En segundo lugar, definimos el test y una regla de decisión del tipo “RECHAZAR si los resultados del test son xxx, ACEPTAR en caso contrario”. En este ejemplo, definimos el siguiente test: “lanzar la moneda 100 veces” y el resultado del test es “número de caras”. La regla de decisión es muy clara, si salen muchas caras o pocas caras RECHAZAR, si salen alrededor de 50 caras, ACEPTAR.  En la jerga del contraste de hipótesis se llama “estadístico” al resultado del test, por lo tanto, en este caso, el estadístico es el número de caras.

### ***Nivel de confianza y región de rechazo***

Finalmente, para determinar exactamente el número de caras que nos llevaría a RECHAR tenemos que añadir un nuevo criterio para determinar el grado de confianza que deseamos tener al RECHAZAR la hipótesis nula. Si para RECHAZAR exigimos un número de caras muy bajo o muy alto, entonces tendremos un “nivel de confianza” muy elevado, en el sentido de que cuando rechazamos podemos estar muy seguros (mucha confianza) de que realmente la moneda no está equilibrada. Por ejemplo, podemos establecer la ”región de rechazo” de 0 a 10 y de 90 a 100. De esta forma, cuando el número de caras es inferior a 10 (incluido) o superior a 90 (incluido), se rechaza la hipótesis nula. Con esta región de rechazo podemos estar muy seguros de que nunca vamos a rechazar una hipótesis nula cuando la moneda sea equilibrada, ya que es muy improbable obtener ese resultado con una moneda equilibrada. De hecho, es prácticamente imposible.  En la gráfica siguiente se puede ver es muy improbable que sacar menos de 32 caras o más de 68 caras.

Normalmente es suficiente con elegir un nivel de confianza del 95%, aunque también se admiten valores de 90% y 99% dependiendo del grado de confianza que queramos tener en hacer lo correcto cuando se rechaza la hipótesis nula. Una vez decidido el nivel de confianza, la región de rechazo está perfectamente definida ya que depende de la distribución de probabilidad del estadístico bajo la hipótesis nula. En nuestro caso, de la distribución de probabilidad del número de caras de una moneda equilibrada.

La significatividad es el complemento del nivel de confianza, es decir, un nivel de confianza del 95% tiene una significatividad del 5%, un nivel de confianza del 90% tiene una significatividad del 10%, etc.


### ***p-valor de un estadístico.***

Otra forma de realizar el mismo análisis sería observar primero el resultado del test y después decidir si se RECHAZA o no la hipótesis nula dependiendo del grado de compatibilidad de la hipótesis con el resultado obtenido. De esta forma no necesitamos definir a priori la región de rechazo. En nuestro ejemplo sería suficiente con lanzar la moneda 100 veces, contar el número de caras y valorar el grado de compatibilidad entre ese resultado con la hipótesis nula. Si el resultado es muy improbable, entonces RECHAZAMOS la hipótesis nula y en caso contrario la ACEPTAMOS.

Para analizar la compatibilidad entre la hipótesis nula y el resultado del test se utiliza el concepto de p-valor, que se define como la probabilidad de obtener ese resultado o un resultado más desfavorable suponiendo que la hipótesis nula es cierta. Cuando la región de rechazo se define en los dos extremos del estadístico, entonces el p-valor se define como la probabilidad en los dos lados del estadístico. En nuestro caso, la región de rechazo incluye los valores bajos o altos del número de caras, es decir, los dos extremos. Por ejemplo, el p-valor de obtener 38 caras se define como la probabilidad de obtener 38 caras o menos más la probabilidad de obtener 62 caras o más. Dicha probabilidad es 0,021, lo que nos llevaría a la conclusión de que podemos RECHAZAR la hipótesis nula “la moneda es equilibrada” con un nivel de confianza del 95% pero NO podemos rechazarla con un nivel de confianza del 99%.

### ***Error tipo I y tipo II.***

Como hemos visto, cuando RECHAZAMOS la hipótesis nula corremos el riesgo de rechazarla equivocadamente, es decir, rechazar cuando es cierta. Esto sucede muy poco y la probabilidad de que suceda es precisamente la significatividad, es decir, el complementario del nivel de confianza. A este tipo de error se le llama error de tipo I.

Pero también podemos cometer el error contrario, es decir, ACEPTAR la hipótesis nula cuando en realidad es falsa. En nuestro caso cometeríamos este error cuando ACEPTAMOS que la moneda es equilibrada, pero en realidad no lo es. A este tipo de error se llama error de tipo II. Se define la “potencia de un test” como 1 menos la probabilidad de cometer un error de tipo II.

Normalmente los errores tipo I y II son antagónicos ya que si minimizamos uno de ellos empeoramos el otro. No obstante, algunos tipos de tests sofisticados permiten definir un estadístico que mejora algo los dos tipos de errores al mismo tiempo. 

En la práctica, es suficiente con fijar el nivel de significatividad o la probabilidad de cometer un error de tipo I y no cuantificar la probabilidad de cometer el error de tipo II

### ***Experimento:***

***Suponemos que tiramos una moneda al aire 100 veces:***

```python
import random
caras=0
cruces=0
for i in range(100):
  moneda=random.choice(["cara", "cruz"])
  if moneda=="cara":
    caras +=1
  elif moneda=="cruz":
    cruces +=1

print("Han salido", caras, "caras, y", cruces,"cruces")
```

Por lo que no podemos rechazar la hipótesis nula $H_0$ a un nivel de confianza del 90%, la moneda está equilibrada.

***Suponemos que tiramos ahora la moneda 100.000 veces:***

```python
import random
caras=0
cruces=0
for i in range(100000):
  moneda=random.choice(["cara", "cruz"])
  if moneda=="cara":
    caras +=1
  elif moneda=="cruz":
    cruces +=1

print("Han salido", caras, "caras, y", cruces,"cruces")
```

Por lo que no podemos rechazar la hipótesis nula $H_0$ del 90%, la moneda está equilibrada.
